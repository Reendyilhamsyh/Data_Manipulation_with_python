```python
#Data Transformation
df = pd.DataFrame({'A': np.random.randint(low=1,high=10,size=100),
                   'B': np.random.randint(low=10,high=100,size=100),
                   'C': np.random.randint(low=20,high=45,size=100)
                   }, )
df
```

```python
#misal kita ingin membuat kolom baru yaitu kolom "D",dengan rumus D = 10*A+(0,5)*B-Akar(C)
#cukup tulis satu fungsi saja dan langsung bisa menghitung seluruh baris di dataframe kita,dengan menggunakan metod .apply
df['D'] = df.apply(lambda x: 10*x['A'] + (0.5)*x['B'] - np.sqrt(x['C']), axis=1)
df
```

```python
#Apply dapat berfungsi sebagai pengulangan untuk memasukkan nilai tiap naris kedalam fungsi custom yang kita buat
#kalau cuma pengen tau rata rata per barisnya tanpa seribet itu menggunakan yang umum saja,nggak usah pake apply lambda
df['rata-rata_abc'] = df[['A','B','C']].mean(axis=1) #axis=1 itu kanan ke kiri (baris),axis=0 itu atas ke bawah/kolom
df
```

```python
#melakukan pemetaan
#misal
#-Jika kolom A bernilai 1-3,beri label 'Kurang'
#-Jika kolom B bernilai 4-7,beri label 'Cukup'
#-Jika kolom C bernilai 8-10,beri label 'Surplus'
#karena kita hanya membutuhkan input satu saja yaitu kolom a,maka tidak perlu menggunakan lambda,cukup dengan
#1.Buat Function
def label(a):
    if a<4:
        return 'Kurang'
    elif a<8:
        return 'Cukup'
    else:
        return 'Surplus'
#
```

```python
df['label_A'] = df['A'].apply(label)
df
```

```python
#Data Standardization
#Standarisasi adalah sebuah proses untuk membuat distribusi data itu memiliki rata rata 0 dan standar deviasi 1
#Penggunaannya itu ketika menggunakan algoritma Machine Learning yang distance base,seperti : KNN,SVM,K-Means,dll
#Hindari penggunaan saat menggunakan logistic reggression dan tree based model
#Berikut contoh melakukan standarisasi menggunakan library sklearn
#1.Panggil library dari sklearn
from sklearn import preprocessing
```

```python
#import data yang kita punya sebelum standarisasi
iris
```

```python
#Mulai standarisasi menggunakan method atau fungsi bawaan sklearn di dalam parameter dataframe kita
standardized_iris = pd.DataFrame(preprocessing.scale(iris[['sepal_length',
                                       'sepal_width',
                                       'petal_length',
                                       'petal_width']]))
standardized_iris
#bisa dilihat kolom berubah dengan skala yang telah terstandarisasi.Perbedaan dengan min,max scaller
#yaitu mematikan angka minimum adalah 0 dan maksimum 1,sedangkan standarization terdapat
#angka negatif,yang dimana berfokus untuk rata rata mendekati 0
```

```python
#Mengecek data yang sudah distandarisasi dengan menghitung menggunakan method .mean apakah mendekati 0?
standardized_iris.mean(axis=0)
#Makna Nilai Kecil:
#-Nilai mean yang sangat kecil (mendekati 0) menunjukkan bahwa data telah distandarisasi dengan benar.
#-Dalam konteks standarisasi data, mean setiap fitur (kolom) harus mendekati 0, dan standar deviasi mendekati 1.

#Kesimpulan dari Output:
#Data iris yang telah distandarisasi memiliki rata-rata yang sangat mendekati 0 untuk setiap kolom.
#Hal ini mengindikasikan bahwa proses standarisasi berhasil, sehingga data siap untuk digunakan dalam analisis atau algoritma machine learning yang membutuhkan data yang distandarisasi.
#Secara keseluruhan, output tersebut menunjukkan bahwa standarisasi data iris telah dilakukan dengan baik dan hasilnya sesuai dengan harapan.
```

```python
#mengecak data yang distandarisasi dengan menghitung std deviasi menggunakan method .std yang apakah mendekati 1?
standardized_iris.std(axis=0)
#Makna Nilai Standar Deviasi:
#-Nilai standar deviasi yang mendekati 1 menunjukkan bahwa data telah distandarisasi dengan benar.
#-Dalam konteks standarisasi data, standar deviasi setiap fitur (kolom) harus mendekati 1.

#-Kesimpulan dari Output:
#-Data iris yang telah distandarisasi memiliki standar deviasi yang sangat mendekati 1 untuk setiap kolom.
#-Hal ini mengindikasikan bahwa proses standarisasi berhasil, sehingga data siap untuk digunakan dalam analisis atau algoritma machine learning yang membutuhkan data yang distandarisasi.
#-Secara keseluruhan, output tersebut menunjukkan bahwa standarisasi data iris telah dilakukan dengan baik dan hasilnya sesuai dengan harapan, dengan standar deviasi setiap fitur mendekati 1.
```
